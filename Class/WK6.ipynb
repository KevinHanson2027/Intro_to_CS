```python
import random as r

num = r.randint(1500, 2500)
num_str = str(num)

all_digits_even = True
for digit in num_str:
    if int(digit) % 2 != 0:
        all_digits_even = False
        break

if all_digits_even:
    print(f"Yes! {num} is entirely even")
else:
    print("Not all digits are even")
```
