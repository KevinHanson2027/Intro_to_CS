### Write a Python script that performs the following steps:
- Given the string "I am an enchanter. There are some who call me...Tim.", remove all spaces from the string.
- Reverse the string and print it character by character.
- For each character, alternate between printing it in uppercase and lowercase, starting with uppercase for the first character in the reversed string.

### Requirements:
- The string should be processed without spaces.
- The output should begin with the first character in uppercase, and then alternate between lowercase and uppercase for the following characters.
- The characters are printed directly next to each other without newlines between them.

---


```python
# Task 1
var = "I am an enchanter. There are some who call me...Tim."
var2 = 0

while var2 < len(var):
    if var[var2] != " " and var[var2] != " ":
        print(var[var2], end="")
    var2 += 1
```

```python
# Task 2
var = "I am an enchanter. There are some who call me...Tim."
var2 = 0

for i in range(len(var) - 1, -1, -1):
    print(var[i], end='')
```

```python
# Task 3
var = "I am an enchanter. There are some who call me...Tim."
var2 = 0

for i in range(len(var) - 1, -1, -1):
    if i == len(var) - 1:
        print(var[i].upper(), end="")
    else:
        print(var[i].lower() 
              if i % 2 == 0 
              else var[i].upper(), end="")
```
